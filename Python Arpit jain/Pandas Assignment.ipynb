```python
##Which zip code has the costliest house?


costliest_zipcode = df.loc[df['price'].idxmax()]['zipcode']

print(f'The costliest house is in the zip code: {costliest_zipcode}')
```

```python
##Which zip code area has the cheapest house?

min_price_row = df[df['price'] == df['price'].min()]
cheapest_zip_code = min_price_row['zipcode'].values[0]
print(cheapest_zip_code)
```

```python
##Which area has the biggest sqft_living?
max_sqft_living_row = df[df['sqft_living'] == df['sqft_living'].max()]

max_sqft_living_row
```
